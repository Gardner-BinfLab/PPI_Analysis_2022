```python
import os
import pickle
from glob import glob
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from upsetplot import generate_counts, plot, UpSet
from protlearn.features import aaindex1
from protlearn.preprocessing import remove_unnatural
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, average_precision_score
from sklearn.model_selection import train_test_split
from scipy.stats import spearmanr


def fasta_reader(file):
    '''Converts .fasta to a pandas dataframe with accession as index
    and sequence in a column 'sequence'
    '''
    fasta_df = pd.read_csv(file, sep='>', lineterminator='>', header=None)
    fasta_df[['Accession', 'Sequence']] = fasta_df[0].str.split('\n', 1, \
                                        expand=True)
    fasta_df['Accession'] = fasta_df['Accession']
    fasta_df['Sequence'] = fasta_df['Sequence'].replace('\n', '', regex=True).\
                            astype(str).str.upper().replace('U', 'C')
    total_seq = fasta_df.shape[0]
    fasta_df.drop(0, axis=1, inplace=True)
    fasta_df = fasta_df[fasta_df.Sequence != '']
    fasta_df = fasta_df[fasta_df.Sequence != 'NONE']
    final_df = fasta_df.dropna()
    remained_seq = final_df.shape[0]
    if total_seq != remained_seq:
        print("{} sequences were removed due to inconsistencies in"
                      "provided file.".format(total_seq-remained_seq))
    return final_df


# https://stackoverflow.com/a/66893520
# https://stackoverflow.com/a/50690729
def reg_coef(x,y,label=None,color=None,hue=None,**kwargs):
    ax = plt.gca()
    corr_r,p = spearmanr(x,y)
    corr_text = f"{corr_r:2.2f}".replace("0.", ".")
    ax.set_axis_off()
    marker_size = abs(corr_r) * 10000
    ax.scatter([.5], [.5], marker_size, [corr_r], alpha=0.6, cmap="coolwarm",
               vmin=-1, vmax=1, transform=ax.transAxes)
    font_size = abs(corr_r) * 40 + 5
    ax.annotate(corr_text, [.5, .5,],  xycoords="axes fraction",
                ha='center', va='center', fontsize=font_size)
```

### Built interaction type-specific classifiers

#### STRING protein.actions

```python
dbs = pd.read_pickle('dbs.pkl.gz')
dp = pd.read_pickle('string_act.pkl.gz').drop_duplicates('STRING', keep=False)

df_int = pd.read_pickle('bioplex_interactions_aaindex1.pkl.gz')
df_int['Pairs'] = df_int[['UniprotA','UniprotB']].values.tolist()
df_int['Pairs'] = df_int.Pairs.apply(lambda x: '_'.join(sorted([x[0], x[1]])))
df_int = df_int.drop_duplicates('Pairs')
df_int.rename(columns={'Interaction':'Label'}, inplace=True)

dn = pd.concat([dbs, df_int[df_int.Label==0]])[['Pairs','Label','aaind']].drop_duplicates('Pairs', keep=False)
dn = dn[dn.Label==0]

mode = list(dp['mode'].unique())

size = []
clfs = []
for i,v in enumerate(mode):
    dd = pd.concat([dp[dp['mode']==v], dn]).sample(frac=1,random_state=12345)
    clf = RandomForestClassifier(random_state=12345, n_jobs=-1)
    clf.fit(dd.aaind.tolist(), dd.Label.tolist()).set_params(n_jobs=1)
    clfs.append(clf)
    size.append([v, len(dd[dd.Label==1]), len(dd[dd.Label==0])])
    print('clf # {}.'.format(i), end='\r')
with open('../../clf/final_string_mode_clfs.pickle', 'wb') as handle:
    pickle.dump(clfs, handle)

size = pd.DataFrame(size) #.sort_values([1], ascending=False)
size
```

#### signor interaction types

```python
dbs = pd.read_pickle('dbs.pkl.gz')
signor = pd.read_csv('../../ref/signor/Jan2022_release.txt.gz', sep='\t')
signor = signor[(signor.DATABASEA=='UNIPROT') & (signor.DATABASEB=='UNIPROT')].copy()
signor['Pairs'] = signor[['IDA','IDB']].values.tolist()
signor['Pairs'] = signor.Pairs.apply(lambda x: '_'.join(sorted(x)))

df_int = pd.read_pickle('bioplex_interactions_aaindex1.pkl.gz')
df_int['Pairs'] = df_int[['UniprotA','UniprotB']].values.tolist()
df_int['Pairs'] = df_int.Pairs.apply(lambda x: '_'.join(sorted([x[0], x[1]])))
df_int = df_int.drop_duplicates('Pairs')
df_int.rename(columns={'Interaction':'Label'}, inplace=True)


dp = []
for i in signor.value_counts('MECHANISM').head(10).index:
    pp = pd.merge(dbs, signor[signor.MECHANISM==i][['Pairs']], on='Pairs')
    pp['MECHANISM'] = i
    dp.append(pp)   
dp = pd.concat(dp).drop_duplicates(['Pairs','MECHANISM'])
dp['Label'] = 1

dn = pd.concat([dbs, df_int[df_int.Label==0]])[['Pairs','Label','aaind']].drop_duplicates('Pairs', keep=False)
dn = dn[dn.Label==0]

size = []
clfs = []
for i,v in enumerate(signor.value_counts('MECHANISM').head(10).index):
    dd = pd.concat([dp[dp.MECHANISM==v], dn]).sample(frac=1,random_state=12345)
    clf = RandomForestClassifier(random_state=12345, n_jobs=-1)
    clf.fit(dd.aaind.tolist(), dd.Label.tolist()).set_params(n_jobs=1)
    clfs.append(clf)
    size.append([v, len(dd[dd.Label==1]), len(dd[dd.Label==0])])
    print('clf # {}.'.format(i), end='\r')
with open('../../clf/final_signor_clfs.pickle', 'wb') as handle:
    pickle.dump(clfs, handle)
    
size = pd.DataFrame(size)
size #.sort_values([1], ascending=False)
```

#### Scatterplot matrix of feature importances

```python
! wget https://www.genome.jp/aaindex/AAindex/Appendix
! wget https://www.genome.jp/aaindex/AAindex/list_of_indices
```

```python
with open('Appendix', encoding = "ISO-8859-1") as f:
    lines = f.readlines()
    
a = pd.DataFrame(lines)
a['Class'] = a[0].str.split().apply(lambda x: x[0] if len(x)>0 else np.nan)
a['AAindex'] = a[0].str.split().apply(lambda x: x[1] if len(x)>0 else np.nan)
a = a.dropna()


with open('list_of_indices', encoding = "ISO-8859-1") as f:
    lines = f.readlines()
    
l = pd.DataFrame(lines).loc[5:]
l['AAindex'] = l[0].str.split().apply(lambda x: x[0] if len(x)>0 else np.nan)
l[0] = l[0].str.replace(r'[A-Z]+[0-9]+\s+','', regex=True).str.strip()
l.columns = ['Description','AAindex']
```

```python
aai, ind = aaindex1('AGVWPY')

dp = pd.read_pickle('string_act.pkl.gz').drop_duplicates('STRING', keep=False)
mode = list(dp['mode'].unique())

clfs = pd.read_pickle('../../clf/final_string_mode_clfs.pickle')

gini = []
for i in clfs:    
    importances = i.feature_importances_
    std = np.std([tree.feature_importances_ for tree in i.estimators_], axis=0)
    gi = pd.DataFrame(importances, index=ind)
    gi['std'] = std
    gi.reset_index(inplace=True)
    gi.columns = ['AAindex','Gini','Stdev']
    gini.append(gi)
    
for i,v in enumerate(mode):
    gini[i]['mode'] = v

gini = pd.concat(gini)

gini_aaind = pd.merge(gini, a, on='AAindex')
gini_aaind['Group'] = gini_aaind[0].str.split().apply(lambda x: x[0][0])
gini = pd.concat([gini_aaind, gini]).drop_duplicates(['AAindex','Gini','Stdev','mode'])
gini = gini.fillna('O')

ginit = gini.pivot(index='mode', columns=['AAindex','Group'], values='Gini').T
ginit = ginit[mode]
ginit = ginit.reset_index()

props = {'A':'Alpha and turn propensities','B':'beta propensity',
         'C':'Composition','H':'Hydrophobicity',
         'P':'Physicochemical properties','O':'Other properties'}

ginit['Group'] = ginit.Group.apply(lambda x: props[x])


# sns.set(font_scale=1.3) #style='white', 
g = sns.PairGrid(ginit, diag_sharey=False, hue='Group')
# g.map_lower(sns.regplot, scatter=False, lowess=True, ci=False, line_kws={'color': 'black'})
g.map_lower(sns.scatterplot)
g.map_diag(sns.kdeplot)
g.map_upper(reg_coef)
g.add_legend()
plt.savefig('../../fig/final_string_mode_clfs_corrmatrix.pdf')
```

```python
gini.sort_values('Gini').drop_duplicates('mode')
```

```python
top10 = []

for i,v in enumerate(mode):
    top10.append(gini[gini['mode']==v].sort_values('Gini').head(10))
    
top10 = pd.concat(top10).drop('Class', axis=1)
top10['Group'] = top10.Group.apply(lambda x: props[x])
top10 = pd.merge(l[['AAindex','Description']], top10, on='AAindex').drop([0], axis=1)
top10 = top10.set_index('mode').loc[mode].reset_index()
top10 = top10[['mode','AAindex','Description','Group','Gini','Stdev']]
```
